Chi Square

Research Question:

Is there relation of crater depth and depend on crater diameter image?

Null Hypothesis (H0):

Crater depth is not associated with crater diameter image.

Alternative Hypothesis (H1):

Crater depth is some times associated with crater diameter image.



I use a main variable Diam Circle image, then i test the independence. I know that there is a correlation between Diagram image and depth. 
Then i test it again but with ejecta and the latitud.
The data set is not categorical in any way so the results only give some how an aproximation, neverless the infomation of the tipe of craters ar very interesting.

```python
%matplotlib inline
import pandas
import numpy
import scipy.stats
import seaborn
import matplotlib.pyplot as plt
data = pandas.read_csv("crateres.csv")


M = "DIAM_CIRCLE_IMAGE"
S = "NUMBER_LAYERS"


sub1 = data[(data["DIAM_CIRCLE_IMAGE"]>=1) & (data["DIAM_CIRCLE_IMAGE"]<=85)]
            
sub2 = sub1.copy()


ct1=pandas.crosstab(sub2['DIAM_CIRCLE_IMAGE'], sub2['MORPHOLOGY_EJECTA_1'])
print (ct1)

colsum=ct1.sum(axis=0)
colpct=ct1/colsum
print(colpct)


print ('chi-square value, p value, expected counts')
cs1= scipy.stats.chi2_contingency(ct1)
print (cs1)
```

Crosstabs with longitud and Ejecta morphology

And the Factorsplot

```python
ct2=pandas.crosstab(sub2['LONGITUDE_CIRCLE_IMAGE'], sub2['MORPHOLOGY_EJECTA_1'])
#print (ct2)

colsum=ct2.sum(axis=0)
colpct=ct2/colsum
print(colpct)


print ('chi-square value, p value, expected counts')
cs1= scipy.stats.chi2_contingency(ct1)
print (cs1)




seaborn.factorplot(x="LONGITUDE_CIRCLE_IMAGE", y="MORPHOLOGY_EJECTA_1", data=sub2, kind="bar", ci=None,size=8,aspect=2)
plt.xlabel('Longitud Mars')
plt.ylabel('Ejecta Type')



seaborn.factorplot(x="DIAM_CIRCLE_IMAGE", y="MORPHOLOGY_EJECTA_1", data=sub2, kind="bar", ci=None,size=8,aspect=2)
plt.xlabel('Diagram circle Image')
plt.ylabel('Ejecta Type')



```
